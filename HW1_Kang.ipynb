```python
import numpy as np
import pandas as pd

# Read the excel sheet to pandas dataframe
HIV = pd.read_excel("HIV.xlsx", sheetname=0)
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
```

```python
from bokeh.models import ColumnDataSource

###Change original dataframe to dictionary with three columns: country, year, prevalence
country = np.array(HIV[HIV.columns.values[0]])
country1 = np.repeat(country, HIV.shape[1]-1)


year = HIV.columns.values[1:]
year = year.astype(int)

year1 = np.tile(year, HIV.shape[0])

hiv = HIV[HIV.columns.values[1:]]
hiv1 = hiv.values.flatten()

source = ColumnDataSource(data={
    "country": country1,
    "year": year1,
    'hiv': hiv1})
print(len(country1),len(year1),len(hiv1))
```

```python
from bokeh.models import HoverTool
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.palettes import Viridis256

source = ColumnDataSource(data={
    "country": country1,
    "countryint": np.tile(range(HIV.shape[0]),HIV.shape[1]-1),
    "year": year1,
    'hiv': hiv1})

hover = HoverTool(
        tooltips=[
            ("Country", "@country"),
            ("Year", "@year"),
            ("Prevalence", "@hiv"),
        ]
    )

color_mapper = LinearColorMapper(palette=Viridis256, low=1, high=HIV.shape[0])

p = figure(x_axis_label='Year', y_axis_label='Prevalence %', tools=[hover], toolbar_location='above',title="HIV prevalence")
p.circle(x='year', y='hiv', color={'field': "countryint", 'transform': color_mapper}, size=10, alpha=0.6, source=source)

color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0,0), title='Country')
p.add_layout(color_bar, 'right')

show(p)
```
